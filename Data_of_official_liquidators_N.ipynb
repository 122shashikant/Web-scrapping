```python
#importing the libraries needed 
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs
from time import sleep
from random import randint
from urllib.request import urlopen as uReq
```

```python
url='https://www.mca.gov.in/content/mca/global/en/contact-us/official-liquidators.html'
```

```python
uClient = uReq(url)
```

```python
uClient 
```

```python
page=uClient.read()
```

```python
uClient.close()
```

```python
page_html = bs(page, "html.parser")
```

```python
page_html 
```

```python
company=page_html.findAll('tr',{'class':'view table-values dark-theme diff-designation'})
```

```python
company[0].text.strip().split('\n')
```

```python
#declaring the list of empty variables, So that we can append the data overall
Names=[]
Designation=[]
Contact_Details=[]
Address=[]
#the whole core of the script
for j in range(len(company)):
    Names.append(company[j].text.strip().split('\n')[0])
    Designation.append(company[j].text.strip().split('\n')[1])
    Contact_Details.append(company[j].text.strip().split('\n')[2:-3])
    Address.append(company[j].text.strip().split('\n')[-3:])
#creating a dataframe     
List_of_official_liquidators = pd.DataFrame({ "Names": Names, "Designation" :Designation, "Contact_Details": Contact_Details,"Address": Address,})
List_of_official_liquidators.head()
```

```python
# #saving the data in csv format
List_of_official_liquidators.to_csv('Data_official_liquidators',index=False)
```
